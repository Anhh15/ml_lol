# T2 — Tuning XGBoost (Early Game Objectives)
GridSearchCV trên `t2_train.csv` | cv=5 | scoring=accuracy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

ROOT = Path('../../')
train = pd.read_csv(ROOT / 'data/processed/t2_train.csv')
X_train = train.drop(columns=['winner'])
y_train = train['winner']

sns.set_theme(style='whitegrid', font_scale=1.0)
plt.rcParams['figure.dpi'] = 110
C1, C2 = '#4C82C2', '#E07B54'

print(f'Train: {X_train.shape}')
```

```python
param_grid = {
    'n_estimators':    [100, 300],
    'max_depth':       [3, 5, 7],
    'learning_rate':   [0.05, 0.1, 0.3]
}

xgb = XGBClassifier(
    eval_metric='logloss',
    random_state=42,
    n_jobs=-1,
    verbosity=0,
)

gs = GridSearchCV(xgb, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
gs.fit(X_train, y_train)

print(f'Best params : {gs.best_params_}')
print(f'Best CV Acc : {gs.best_score_:.4f} ({gs.best_score_*100:.2f}%)')
```

```python
# Feature importance từ best model
best_xgb = gs.best_estimator_
fi = pd.Series(best_xgb.feature_importances_, index=X_train.columns)
top15_fi = fi.nlargest(15)

fig, ax = plt.subplots(figsize=(9, 5))
ax.barh(top15_fi.index[::-1], top15_fi.values[::-1], color=C1)
ax.set_xlabel('Feature Importance')
ax.set_title('Top 15 feature quan trọng nhất (XGBoost T2)')
plt.tight_layout()
plt.show()
```

```python
print('=' * 55)
print('KET QUA - XGBOOST (T2)')
print('=' * 55)
print(f'Best CV Accuracy : {gs.best_score_:.4f}  ({gs.best_score_*100:.2f}%)')
print(f'Best params      : {gs.best_params_}')
print()
print('=> Dung cho t2_train_best.py:')
print(f"BEST_MODEL  = 'xgb'")
print(f"BEST_PARAMS = {{")
for k, v in gs.best_params_.items():
    print(f"    '{k}': {repr(v)},")
print(f"    'eval_metric'  : 'logloss',")
print(f"    'random_state' : 42,")
print(f"    'n_jobs'       : -1,")
print(f"    'verbosity'    : 0,")
print(f"}}")
```
